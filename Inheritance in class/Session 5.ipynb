# Inheritance in Python
### All properties from parent class will be transferred to child class

### Types of Inheritance
1.Single Inheritance

2.Multiple Inheritance

3.Hierarhcical Inheritance 

```python
#Single Inheritance example
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def introduce(self):
        print(f'My name is {self.name}, Im {self.age} years old.')
```

```python
class Employee(Person):
    def __init__(self,name,age,company):
        super().__init__(name,age)
        self.company = company
    def work(self):
        super().introduce()
        print(f'I work from {self.company} company.')
```

```python
p1 = Person('kalpana',48)
```

```python
p1.name
```

```python
p1.age
```

```python
p1.introduce()
```

```python
e1 = Employee('Yukta',24,'self')
```

```python
e1.name
```

```python
e1.age
```

```python
e1.company
```

```python
e1.work()
```
